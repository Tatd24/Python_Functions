1. What is the difference between a function and a method in Python?

  Ans: In Python, functions and methods are both blocks of reusable code, but there are key differences between them based on how they are defined and used:

  Function:

*   A function is defined using the def keyword.
*   It stands alone and is not associated with any object or class.

*   You can call a function directly by its name (if in the same scope).
*   Example:
        def greet(name):
            return f"Hello, {name}!"

        print(greet("Alice"))

  Method:

*   A method is also defined using the def keyword, but inside a class.
*   It is associated with an object (instance) of that class.

*   The first parameter is usually self, which refers to the instance calling the method.
*   You call a method on an object.

*   Example:

class Person:

        def greet(self, name):

            return f"Hello, {name}!"

        p = Person()
        print(p.greet("Alice"))

4. What is the purpose of the `return` statement in a Python function?
  
  Ans: The return statement in a Python function is used to:

*  When a function completes its task, return allows it to pass a value (or multiple values) back to the code that called it. This is useful when you want to use the function’s output elsewhere in your program.
*   Once the return statement is executed, the function stops running, and control goes back to the caller. Any code written after the return statement will not be executed.


*  Example:

        def add(a, b):
            return a + b

6. Explain the concept of generators in Python and how they are defined.

  Ans: Generators are special functions that yield values one at a time using the yield keyword, instead of returning them all at once.



*   Define a Generator:
        def count_up_to(n):
            i = 1
            while i <= n:
                yield i
                i += 1
*   Key Points:

  Use yield instead of return

  Memory efficient

  Generates values lazily

  Can be looped with for or next()

  Example:
  
        gen = count_up_to(3)
        print(next(gen))  # 1






7. What are the advantages of using generators over regular functions?

  Ans: Advantages of Using Generators over Regular Functions:


*   Memory Efficient: They yield one item at a time, avoiding storing the whole sequence in memory.
*   Lazy Evaluation: Values are generated only when needed, improving performance.

*   Infinite Sequences:Ideal for producing infinite data streams (like sensor readings).
*   Cleaner Code: Simpler than managing state with loops and lists.

  Example:

         def gen_numbers():
             for i in range(1000000):
                 yield i

  This uses much less memory than returning a list of a million numbers.

1. Write a Python function that takes a list of numbers as input and returns the sum of all even numbers in
the list

```python
def sum_of_even_numbers(numbers):
    return sum(num for num in numbers if num % 2 == 0)

# Example usage:
my_list = [1, 2, 3, 4, 5, 6]
print("Sum of even numbers:", sum_of_even_numbers(my_list))
```

4. Write a Python function that checks if a given number is prime or not from 1 to 200.

```python
def is_prime(num):
    if num <= 1:
        return False
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
    return True
n = int(input("Enter a number between 1 and 200: "))

if 1 <= n <= 200:
    if is_prime(n):
        print(n, "is a prime number")
    else:
        print(n, "is not a prime number")
else:
    print("Please enter a number between 1 and 200")
```

5. Create an iterator class in Python that generates the Fibonacci sequence up to a specified number of
terms.

```python
class FibonacciIterator:
    def __init__(self, n_terms):
        self.n_terms = n_terms   # total terms to generate
        self.count = 0           # counter for generated terms
        self.a, self.b = 0, 1    # starting values of Fibonacci

    def __iter__(self):
        return self

    def __next__(self):
        if self.count >= self.n_terms:
            raise StopIteration   # stop when terms are finished
        if self.count == 0:
            self.count += 1
            return 0
        elif self.count == 1:
            self.count += 1
            return 1
        else:
            self.a, self.b = self.b, self.a + self.b
            self.count += 1
            return self.b
n = int(input("Enter number of terms for Fibonacci sequence: "))
fib = FibonacciIterator(n)

print("Fibonacci sequence:")
for num in fib:
    print(num, end=" ")
```

6.  Write a generator function in Python that yields the powers of 2 up to a given exponent.

```python
def powers_of_two(limit):
    for exp in range(limit + 1):
        yield 2 ** exp
n = int(input("Enter the exponent: "))
print(f"Powers of 2 up to 2^{n}:")

for value in powers_of_two(n):
    print(value, end=" ")
```

7. Implement a generator function that reads a file line by line and yields each line as a string.

```python
def read_file_line_by_line(file_path):
    with open(file_path, 'r') as file:
        for line in file:
            yield line.rstrip('\n')  # Remove trailing newline character

# Example usage:
# Assuming "example.txt" is your file
for line in read_file_line_by_line("example.txt"):
    print(line)
```

8. Use a lambda function in Python to sort a list of tuples based on the second element of each tuple.

```python
# Sample list of tuples
data = [(1, 5), (3, 1), (4, 6), (2, 3)]

# Sort by second element using lambda
sorted_data = sorted(data, key=lambda x: x[1])

print("Original list:", data)
print("Sorted list (by second element):", sorted_data)
```

9. Write a Python program that uses `map()` to convert a list of temperatures from Celsius to Fahrenheit

```python
# List of temperatures in Celsius
celsius_temps = [0, 20, 37, 100]

# Function to convert Celsius to Fahrenheit
def c_to_f(c):
    return (c * 9/5) + 32

# Use map() to apply conversion to each temperature
fahrenheit_temps = list(map(c_to_f, celsius_temps))

# Output the result
print("Fahrenheit temperatures:", fahrenheit_temps)
```

10. Create a Python program that uses filter() to remove all the vowels from a given string.

```python
def remove_vowels(input_str):
    vowels = 'aeiouAEIOU'
    return ''.join(filter(lambda ch: ch not in vowels, input_str))

# Example usage
text = "Hello, World!"
result = remove_vowels(text)
print("String without vowels:", result)
```
